```python
from selenium import webdriver
import time 
```

```python
driver=webdriver.Chrome("D:\chrome-driver\chromedriver")
```

```python
l=[]
with open("../API/datahashes/exe.txt",'r') as reader:
    l.append(reader.read())
```

```python
exe=l[0].split("\n")
```

```python
exe
```

```python
d=[]
for i in exe[:25]:
    driver.get(f"https://www.virustotal.com/gui/file/{i}/details")
    time.sleep(2)
    x=driver.find_elements_by_xpath("//body/vt-ui-shell//div")
    d.append(x[1].text.split("\n"))
```

```python
d
```

```python
det=[]
```

```python
for data in d:
    details=dict()
    try:
        tb=data.index("Basic Properties")
        '''print(tb)
        print("SHA-256",data[data.index("SHA-256")+1])
        print("File Type",data[data.index('File type')+1])
        print("File Size",data[data.index('File size')+1])
        print("Magic",data[data.index('Magic')+1])'''
        
        details.update({data[tb]:{"SHA256":data[data.index("SHA-256")+1],"Magic":data[data.index('Magic')+1],"File Type":data[data.index('File type')+1],"File Size":data[data.index('File size')+1]}})
        #print(details)
    except:
       
        tb=-1
    try:
        th=data.index("History")
        print(data[data.index('Last Analysis')+1])
        details.update({"History":{"Creation Time":data[data.index('Creation Time')+1],"Last Analysis":data[data.index('Last Analysis')+1]}})
    except:
        th=-1
    try:
        tn=data.index("Names")
        try:
            m=data.index("Signature Info")
        except:
            try:
                m=data.index("Portable Executable Info")
                print(m)
            except:
                m=len(data)-1
        details.update({"Names":{"Number of names":m-tn-1}})
    except:
        tn=-1
    try:
        ts=data.index("Signature Info")
        details.update({"Signature Info":{"Present":1}})
    except:
        ts=-1
        details.update({"Signature Info":{"Present":-1}})
    try:
        tp=data.index("Portable Executable Info")
        details.update({data[tp]:{"Headers":{"Target":data[data.index('Target Machine')+1],"Entry Point":data[data.index('Entry Point')+1]}}})
    except:
        tp=-1
    try:
        tsec=data.index("Sections")
        i=tsec+2
        m,mx,avg,c=100000.0,0.0,0.0,0
        diff=[]
        fl=data[tsec+1].split()
        tzz=[]
        #print(fl)
        for j in range(len(fl)-1):
            if(fl[j] in ["Virtual","Raw"] and fl[j+1] in ["Size", "Address"]):
                tzz.append(fl[j]+" "+fl[j+1])
                
            elif(fl[j] in ["Size","Address"]):
                pass
            else:
                tzz.append(fl[j])
                
        
        if fl[-1] != tzz[-1][len(tzz[-1])-len(fl[-1]):]:
            tzz.append(fl[-1])
        fl=tzz.copy()
        print(fl)
        
        
        print(len(data[i].split()))
        while(len(fl)==len(data[i].split())):
            try:
                
                ui=fl.index("Virtual Size")
                
                us=fl.index("Raw Size")
                #print(fl[ui],fl[us])
                diff.append(abs(int(data[i].split()[2])-int(data[i].split()[3])))
                
            except:
                print("err in diff")
            #print(float(data[i].split()[4]))
            try:
                dl=fl.index("Entropy")
                if(float(data[i].split()[4])>mx):
                    mx=float(data[i].split()[4])
                avg+=float(data[i].split()[4])
                if(float(data[i].split()[4])<m):
                    m=float(data[i].split()[4])
            except:
                print("missing entropy")
            #print(data[i].split())
            #print(data[i])
            c+=1
            i+=1
        if(c!=0):
            avg=avg/c
        #print(avg,mx,m)
        #print(diff)
        details.update({"Sections":{"Entopy":{"Max":mx,"Min":m,"Mean":avg},"diff size":diff}})
    except:
        tsec=-1
    try:
        ti=data.index("Imports")
        try:
            r=data.index("Exports")
        except:
            try:
                r=data.index("Contained Resources By Type")
            except:
                try:
                    r=data.index("Contained Resources By Language")
                except:
                    r=data.index("Contained Resources")
                
        details.update({"Imports":r-ti-1})
    except:
        ti=-1
    try:
        te=data.index("Exports")
        try:
            r=data.index("Exports")
        except:
            try:
                r=data.index("Contained Resources By Type")
            except:
                try:
                    r=data.index("Contained Resources By Language")
                except:
                    r=data.index("Contained Resources")
                
        details.update({"Exports":r-te-1})
    except:
        te=-1
    try:
        tcr=data.index("Contained Resources By Type")
        
    except:
        tcr=-1
    try:
        tcl=data.index("Contained Resources By Language")
        details.update({"Contained Resources By Language":len(data[data.index("Contained Resources By Language")+1:data.index("Contained Resources")])//2})
    except:
        tcl=-1
    try:
        tc=data.index("Contained Resources")
    except:
        tc=-1
    try:
        to=data.index("Overlay")
        details.update({"Overlay":1})
    except:
        to=-1
        details.update({"Overlay":-1})
    det.append(details)
```

```python
det
```

```python
for i in range(len(det)):
    with open(f"./details/det{i}.json",'w') as writer:
        writer.write(str(det[i]))
```
