```python
import requests
import json
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
types = pd.read_csv("data1_malware.csv")
types
```

```python
types.drop("Unnamed: 0",axis="columns",inplace=True)
types
```

```python
link = []
for ha in types.loc[:,"Hash"]:
    
    link.append("https://www.virustotal.com/gui/file/"+str(ha))
link
```

```python
extension = []
```

```python
for c in range(2333,len(link)):
    print(c)
    url = link[c] + "/details"
    print(url)
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(6)
    #element = driver.find_element_by_css_selector('body')
    #time.sleep(5)
    #element.send_keys(Keys.CONTROL+'a')
    #print(driver.find_element_by_xpath("/html/body").text)
    tt = driver.find_element_by_xpath("/html/body").text
    #if(c == 2):
        #print(tt)
        #break
    #print(len(tt))
    l_text = tt.split("\n")
    try:
        t_indx = l_text.index("File type")
        print(l_text[t_indx+1])
        extension.append(l_text[t_indx+1])
    except:
        time.sleep(60)
        driver.close()
        driver.quit()
        break
    driver.close()
```
