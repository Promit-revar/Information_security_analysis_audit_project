```python
import requests
import json
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
types = pd.read_csv("file_types.csv")
types
```

```python
types.drop("Unnamed: 0",axis="columns",inplace=True)
types
```

```python
link = []
for ha in types.loc[:,"hash(SHA256)"]:
    
    link.append("https://www.virustotal.com/gui/file/"+str(ha))
link
```

```python
types["Vt_link"] = link
types
```

```python
extension = []
```

```python
c = 1
for url in link:
    print(c)
    url = url + "/details"
    print(url)
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(6)
    #element = driver.find_element_by_css_selector('body')
    #time.sleep(5)
    #element.send_keys(Keys.CONTROL+'a')
    #print(driver.find_element_by_xpath("/html/body").text)
    tt = driver.find_element_by_xpath("/html/body").text
    l_text = tt.split("\n")
    try:
        t_indx = l_text.index("File type")
        print(l_text[t_indx+1])
        extension.append(l_text[t_indx+1])
    except:
        extension.append("Not Found")
    driver.close()
    c += 1
```

```python
len(extension)
```

```python
types["File_Exten"] = extension
types
```

```python
types.to_csv('file_types.csv')
```
