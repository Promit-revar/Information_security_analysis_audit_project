## Behaviour

```python
s=[]
m=dict()
dirc=['exe','dll','begnin_dll','begnin_exe']
lens=[5517,627,3233,2129]
for j in range(len(dirc)):
    for i in range(lens[j]):
        with open(f"{dirc[j]}/beh{i}.json","r") as reader:
            t=reader.read()
            t=t.replace("'",'"')
            t=json.loads(t)
            m={"SHA":t["SHA"],"DNS Resolutions":t["Network Communication"]['DNS Resolutions'],"http request":t["Network Communication"]['http request']}
            m.update({"Files opened":t['File Actions']['Files opened'],"Files written":t['File Actions']['Files written'],"Files deleted":t['File Actions']['Files deleted'],"Files copied":t['File Actions']['Files copied'],'Files dropped':t['File Actions']['Files dropped']})
            m.update({'Registry Keys opened':t["Registry Actions"]['Registry Keys opened'],'Registry Keys deleted':t["Registry Actions"]['Registry Keys deleted']})
            m.update({'Shell commands':t['Process And Service Actions']['Shell commands'],'Processes created':t['Process And Service Actions']['Processes created'],'Processes terminated':t['Process And Service Actions']['Processes terminated'],'Services opened':t['Process And Service Actions']['Services opened'],'Services created':t['Process And Service Actions']['Services created']})
            m.update({'Mutexes created':t['Synchronization Mechanisms & Signals']['Mutexes created'],'Mutexes opened':t['Synchronization Mechanisms & Signals']['Mutexes opened']})
            if j>1:
                m.update({"Class":"benign"})
            else:
                m.update({"Class":"malware"})
            s.append(m)
         
```

## Details

```python
s=[]
m=dict()
dirc=['exe','dll','benign_dll','benign_exe']
lens=[5517,627,3210,2129]
for j in range(len(dirc)):
    for i in range(lens[j]):
        with open(f"../details/{dirc[j]}/det{i}.json","r") as reader:
            t=reader.read()
            t=t.replace("'",'"')
            t=json.loads(t)
            try:
                m={"SHA":t['Basic Properties']["SHA256"],'Magic':t['Basic Properties']['Magic'],'File Type':t['Basic Properties']['File Type'],'File Size':t['Basic Properties']['File Size']}
            except:
                pass
            try:
                m.update({'Creation Time':t['History']['Creation Time']})
            except KeyError:
                m.update({"Creation Time":None})
            try:
                m.update({'Last Analysis':t['History']['Last Analysis']})
            except:
                m.update({'Last Analysis':None})
            try:
                m.update({'No. of Names':t['Names']['Number of names']})
            except:
                m.update({'No. of Names':0})
            m.update({'Signature Info':t['Signature Info']['Present']})
            try:
                m.update({'Target':t['Portable Executable Info']['Headers']['Target']})
            except KeyError:
                m.update({'Target':None})
            try:
                m.update({'Entry Point':t['Portable Executable Info']['Headers']['Entry Point']})
            except KeyError:
                m.update({'Entry Point':None})
            try:
                m.update({'Max Entropy':t['Sections']['Entopy']['Max'],"Min Entropy":t['Sections']['Entopy']['Min'],"Mean Entropy":t['Sections']['Entopy']['Mean'],"Total Size Diffrence":sum(t['Sections']['diff size'])})
            except KeyError:
                m.update({'Max Entropy':None,"Min Entropy":None,"Mean Entropy":None,"Total Size Diffrence":0})
            try:
                m.update({"Imports":t['Imports']})
            except KeyError:
                m.update({"Imports":None})
            try:
                m.update({'Contained Resources By Language':t['Contained Resources By Language']})
            except KeyError:
                m.update({'Contained Resources By Language':None})
            try:
                m.update({'Overlay Present':t['Overlay']})
            except:
                m.update({'Overlay Present':None})
            if j>1:
                m.update({"Class":"benign"})
            else:
                m.update({"Class":"malware"})
                
            #except KeyError:
            #    print("f")
            s.append(m)
```

```python
det=pd.DataFrame(s)
```
