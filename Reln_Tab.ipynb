```python
driver=webdriver.Chrome()
```

```python
d=[]
for i in dll[4076:len(dll)]:  # start from c
    driver.get(f"https://www.virustotal.com/gui/file/{i}/relations")
    time.sleep(2)
    x=driver.find_elements_by_xpath("//body/vt-ui-shell//div")
    d.append(x[1].text.split("\n"))
```

```python
len(d)
```

```python
c = 4076  #display-1
for i in range(0,len(d)):
    bh = d[i]
    info = {"SHA-256":dll[i],"Contacted Domains":0,"Contacted IP Addresses":0,"Contacted URLs":0,"Execution Parents":0,
            "PE Resource Parents":0,"Bundled Files":0,"PE Resource Children":0} 
    c += 1
    print(c)
    try:
        det = bh.index("DETECTION")
        try:
            chk = bh.index("RELATIONS")
            indx = dict()
            #print(bh)
            try:
                bd = bh.index("Contacted Domains")
                indx.update({"Contacted Domains":bd})
            except:
                bd = -1;
            try:
                bi = bh.index("Contacted IP Addresses")
                indx.update({"Contacted IP Addresses":bi})
            except:
                bi = -1
            try:
                bu = bh.index("Contacted URLs")
                indx.update({"Contacted URLs":bu})
            except:
                bu = -1
            try:
                bep = bh.index("Execution Parents")
                indx.update({"Execution Parents":bep})
            except:
                bep = -1
            try:
                brp = bh.index("PE Resource Parents")
                indx.update({"PE Resource Parents":brp})
            except:
                brp = -1
            try:
                bbf = bh.index("Bundled Files")
                indx.update({"Bundled Files":bbf})
            except:
                bbf = -1
            try:
                brc = bh.index("PE Resource Children")
                indx.update({"PE Resource Children":brc})
            except:
                brc = -1
            try:
                bgs = bh.index("Graph Summary")
                indx.update({"Graph Summary":bgs})
            except:
                bgs = -1

            indx = {k: v for k, v in sorted(indx.items(), key=lambda item: item[1])}
            #print(indx)
            
            for j in range(0,len(indx)-1):
                p_key = indx[list(indx.keys())[j]]
                n_key = indx[list(indx.keys())[j+1]]
                #print(p_key)
                #print(n_key)
                if list(indx.keys())[j] == "PE Resource Children" or list(indx.keys())[j] == "Bundled Files":
                    val = (n_key-p_key-4)//4
                    if val != 1:
                        val -= 1
                else:
                    val = (n_key-p_key-1)//4
                    
                info[list(indx.keys())[j]] = val
            
            print(info)
            print("found")
        except:
            print("skip")
        #print(info)
        behaviour.append(info)
    except:
        print("wait")
```
