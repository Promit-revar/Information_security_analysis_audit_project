```python
for i in range(1):
    try: 
        url = 'https://www.virustotal.com/vtapi/v2/file/report'
        res=[]
        k=count
        s=0
        c=0
        for i in range(k,1083):
            c+=1
            if(c==4):
                s=(s+1)%16
                c=0
            print(md_hash[i])
            params = {'apikey':api_keys[s], 'resource': md_hash[i]}
            response = requests.get(url, params=params)
            #print(response.status_code)
            if(response.status_code>=200 or response.status_code<300):
                    res.append(response.json())
                    count += 1
                    #print(response.json())

            else:
                print("Wait")
                time.sleep(120)
    except:
        print(len(res),response)
        print(count)
        len(res)
        for i in range(len(res)):
            f = open(f"Benign_Json/ob{i+k}.json",'w')
            f.write(json.dumps(res[i]))
            f.close()
    time.sleep(10)
```

```python
ha = []
total = []
posi = []
link = []
extension = []
```

```python
dir_list = os.listdir("Benign_Files")
dir_hash = []
for file in dir_list:
    dir_hash.append(file.split(".")[0])
```

```python
for i in range(1082):
    try:
        with open(f'Benign_Json/ob{i}.json','r') as reader:
            t=json.loads((reader.read()))
            ha.append(t['sha256'])
            total.append(t['total'])
            posi.append(t['positives'])
            link.append(t['permalink'])
            try:
                indx = dir_hash.index(t['sha256'])
                extension.append(dir_list[indx].split(".")[-1])
            except:
                extension.append("")
    except:
        pass
```
